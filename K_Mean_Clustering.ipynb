```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=500,centers=4,random_state=43)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X.shape,y.shape
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4) #by default n_clusters=8
km.fit(X)
```

```python
y_pred= km.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c="red")
```

#Noise Data


```python
X,y=make_blobs(n_samples=500,centers=4,cluster_std=2.0,random_state=43)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4) #by default n_clusters=8
km.fit(X)
y_pred= km.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c="red")
```

```python
km.cluster_centers_
```

```python
error=[]
for k in range(1,20):
  km=KMeans(n_clusters=k)
  km.fit(X)
  error.append(km.inertia_)
```

```python
plt.plot(range(1,20),error)
```

#Fail Cases

```python
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=500,noise=0.15,random_state=43)
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
km=KMeans(n_clusters=2) #by default n_clusters=8
km.fit(X)
y_pred= km.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c="red")
```
